```python
import gymnasium as gym
import numpy as np
import cv2
from vizdoom import DoomGame, Mode, ScreenResolution

class VizDoomBasicEnv(gym.Env):
    def __init__(self):
        super().__init__()

        self.game = DoomGame()
        self.game.load_config("basic.cfg")
        self.game.set_window_visible(False)
        self.game.set_mode(Mode.PLAYER)
        self.game.set_screen_resolution(ScreenResolution.RES_160X120)
        self.game.init()

        # Actions: left, right, shoot
        self.actions = [
            [1, 0, 0],
            [0, 1, 0],
            [0, 0, 1],
        ]

        # Observations are 84x84 grayscale
        self.observation_space = gym.spaces.Box(
            low=0, high=1, shape=(84, 84), dtype=np.float32
        )

        self.action_space = gym.spaces.Discrete(len(self.actions))

    # --------------------
    # Preprocess frame
    # --------------------
    def preprocess(self, frame):
        # frame comes as shape (C, H, W) RGB
        frame = np.transpose(frame, (1, 2, 0))  # → (H, W, C)

        frame = cv2.resize(frame, (84, 84))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frame = frame.astype(np.float32) / 255.0
        return frame

    # --------------------
    # Reset
    # --------------------
    def reset(self, *, seed=None, options=None):
        self.game.new_episode()

        state = self.game.get_state().screen_buffer
        obs = self.preprocess(state)

        return obs, {}

    # --------------------
    # Step
    # --------------------
    def step(self, action):
        reward = self.game.make_action(self.actions[action])

        done = self.game.is_episode_finished()

        if done:
            obs = np.zeros((84, 84), dtype=np.float32)
        else:
            state = self.game.get_state().screen_buffer
            obs = self.preprocess(state)

        info = {}

        return obs, reward, done, False, info
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DQN(nn.Module):
    def __init__(self, action_size):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 8, stride=4)
        self.conv2 = nn.Conv2d(32, 64, 4, stride=2)
        self.conv3 = nn.Conv2d(64, 64, 3, stride=1)
        
        self.fc1 = nn.Linear(7*7*64, 512)
        self.fc2 = nn.Linear(512, action_size)

    def forward(self, x):
        x = F.relu(self.conv1(x))  
        x = F.relu(self.conv2(x))  
        x = F.relu(self.conv3(x))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        return self.fc2(x)
```

```python
from collections import deque
import random

class ReplayBuffer:
    def __init__(self, capacity=50000):
        self.buffer = deque(maxlen=capacity)
        
    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))
        
    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        return (
            np.array(states),
            np.array(actions),
            np.array(rewards, dtype=np.float32),
            np.array(next_states),
            np.array(dones, dtype=np.uint8)
        )
        
    def __len__(self):
        return len(self.buffer)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

action_size = 3  # left, right, shoot

policy_net = DQN(action_size).to(device)
target_net = DQN(action_size).to(device)
target_net.load_state_dict(policy_net.state_dict())

optimizer = torch.optim.Adam(policy_net.parameters(), lr=1e-4)
memory = ReplayBuffer()
```

```python
def select_action(state, epsilon):
    if random.random() < epsilon:
        return np.random.randint(0, action_size)
    
    state = torch.FloatTensor(state).unsqueeze(0).unsqueeze(0).to(device)
    with torch.no_grad():
        q_values = policy_net(state)
    return q_values.argmax().item()
```

```python
def train_step(batch_size=32, gamma=0.99):
    if len(memory) < batch_size:
        return
    
    states, actions, rewards, next_states, dones = memory.sample(batch_size)

    states = torch.FloatTensor(states).unsqueeze(1).to(device)
    next_states = torch.FloatTensor(next_states).unsqueeze(1).to(device)
    actions = torch.LongTensor(actions).to(device)
    rewards = torch.FloatTensor(rewards).to(device)
    dones = torch.FloatTensor(dones).to(device)

    q_values = policy_net(states).gather(1, actions.unsqueeze(1)).squeeze(1)

    with torch.no_grad():
        next_q = target_net(next_states).max(1)[0]
        target = rewards + gamma * next_q * (1 - dones)

    loss = F.mse_loss(q_values, target)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
import numpy as np
import matplotlib.pyplot as plt

episode_rewards = []          # total reward per episode
moving_avg_rewards = []       # smoothed reward
window = 20                   # moving-average window
```

```python
import cv2
import os
import numpy as np

def record_episode(env, policy_net, episode_num, fps=30, max_frames=2000):
    """
    Plays one episode with the current policy_net and records it as MP4.
    """
    os.makedirs("videos", exist_ok=True)
    filename = f"videos/episode_{episode_num}.mp4"

    state, info = env.reset()

    # Create video writer
    height, width = 84, 84
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(filename, fourcc, fps, (width, height))

    done = False
    frame_count = 0

    while not done and frame_count < max_frames:
        # Convert frame to uint8 image (0–255)
        frame_img = (state * 255).astype(np.uint8)

        # VizDoom is grayscale → convert to BGR for MP4
        frame_img = cv2.cvtColor(frame_img, cv2.COLOR_GRAY2BGR)

        out.write(frame_img)

        with torch.no_grad():
            state_t = torch.tensor(state, dtype=torch.float32).unsqueeze(0).unsqueeze(0).to(device)
            q_values = policy_net(state_t)
            action = torch.argmax(q_values, dim=1).item()

        next_state, reward, done, truncated, info = env.step(action)
        state = next_state
        frame_count += 1

    out.release()
    print(f"🎥 Video saved: {filename}")
```

```python
num_episodes = 3000
epsilon = 1.0
epsilon_min = 0.05
epsilon_decay = 0.999

for episode in range(num_episodes):
    state, info = env.reset()
    total_reward = 0
    
    done = False
    while not done:
        action = select_action(state, epsilon)
        next_state, reward, done, truncated, info = env.step(action)

        memory.push(state, action, reward, next_state, done)
        
        state = next_state
        total_reward += reward

        train_step(batch_size=32)

    # Update target network
    if episode % 20 == 0:
        target_net.load_state_dict(policy_net.state_dict())
    
    # Decay epsilon
    epsilon = max(epsilon_min, epsilon * epsilon_decay)

    # After finishing an episode:
    episode_rewards.append(total_reward)
    
    if len(episode_rewards) >= window:
        moving_avg = np.mean(episode_rewards[-window:])
    else:
        moving_avg = np.mean(episode_rewards)
    
    moving_avg_rewards.append(moving_avg)
    
    print(f"Episode {episode}  Total Reward: {total_reward}  | Moving Avg({window}): {moving_avg:.2f}")

    # Record video every X episodes
    if episode % 100 == 0 and episode > 0:
        record_episode(env, policy_net, episode)
```

```python
plt.figure(figsize=(12,5))
plt.plot(episode_rewards[:3000], label='Episode Reward')
plt.plot(moving_avg_rewards[:3000], label=f'Moving Average ({window})')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Training Reward')
plt.legend()
plt.grid()
plt.show()
```

```python
def evaluate_agent(num_episodes=10):
    total_rewards = []

    for _ in range(num_episodes):
        state, info = env.reset()     # state shape: (84, 84)
        done = False
        ep_reward = 0

        # Add channel dimension: (1, 84, 84)
        state = np.expand_dims(state, axis=0)

        while not done:
            state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)

            with torch.no_grad():
                q_values = policy_net(state_tensor)
                action = q_values.argmax(dim=1).item()

            next_state, reward, terminated, truncated, info = env.step(action)
            ep_reward += reward

            # Add channel dimension again
            next_state = np.expand_dims(next_state, axis=0)

            state = next_state
            done = terminated or truncated

        total_rewards.append(ep_reward)

    print("Average Reward:", sum(total_rewards) / len(total_rewards))
    print("All Rewards:", total_rewards)

evaluate_agent()
```
